## 1) Teniendo como base el siguiente codigo





```python
# Estado inicial ingresado por usuario
saldo = float(input("Ingrese el saldo inicial: "))
deposito = float(input("Ingrese el monto a depositar: "))
retiro = float(input("Ingrese el monto a retirar: "))

# Actualización del estado
saldo = saldo + deposito
saldo = saldo - retiro

# Interés mensual
interes = 0.02
saldo_final = saldo * (1 + interes)

# Salida
print(f"Saldo final con interés: ${saldo_final:.2f}")
```

##¿Qué representa el estado inicial y el estado final del programa?

* Estado inicial:
Es el conjunto de valores que tienen las variables antes de ejecutar cualquier instrucción (saldo, depósito, retiro). Representa la memoria en el momento de iniciar el programa.

* Estado final:
Es el conjunto de valores que tienen las variables después de ejecutar todas las instrucciones del programa (saldo actualizado con depósitos, retiros e interés).

## Explica cómo cada instrucción afecta la memoria

* Asignación (saldo = ...)
Se crea o actualiza una variable en memoria.

* Operadores aritméticos (saldo + deposito)
Se calcula un nuevo valor y Python lo guarda en memoria.

* Entrada por teclado (input())
El valor ingresado por el usuario se almacena en memoria como texto, luego float() lo convierte a número.

* Impresión (print())
No cambia la memoria, solo muestra información al usuario.

## 3) Teniendo como base el siguiente codigo


```python
inventario = [
    {"nombre": "Pan", "precio": 1500, "stock": 30},
    {"nombre": "Leche", "precio": 3500, "stock": 15},
    {"nombre": "Café", "precio": 8000, "stock": 10},
]

print("Primer producto:", inventario[0]["nombre"])

inventario.append({"nombre": "Huevos", "precio": 12000, "stock": 20})

# Disminuir stock de Leche
for p in inventario:
    if p["nombre"] == "Leche":
        p["stock"] -= 2

# Valor total del inventario
valor_total = sum(p["precio"] * p["stock"] for p in inventario)
print("Valor total inventario:", valor_total)

print("Stock menor a 10:")
for p in inventario:
    if p["stock"] < 10:
        print(p)

for p in inventario:
    p["precio"] *= 1.10
    p["precio"] = round(p["precio"], 2)

buscar = input("Ingrese producto a buscar: ")
encontrado = next((p for p in inventario if p["nombre"].lower() == buscar.lower()), None)

if encontrado:
    print("Producto encontrado:", encontrado)
else:
    print("No existe en el inventario.")

# Eliminar productos con stock 0
inventario = [p for p in inventario if p["stock"] > 0]

# Ordenar por precio ascendente
inventario.sort(key=lambda x: x["precio"])

print("Inventario final:", inventario)
```

## 4) Teniendo en cuenta el siguiente codigo

```python
# Tupla de rectángulo
rectangulo = (5, 3)

ancho, alto = rectangulo
area = ancho * alto
perimetro = 2 * (ancho + alto)

print("Área:", area)
print("Perímetro:", perimetro)

# Intento de cambiar ancho (esto da error)
# rectangulo[0] = 10  # TypeError

# ¿Por qué? → Las tuplas son inmutables.

# Distancia total recorrida
import math

posiciones = [(0, 0), (1, 2), (3, 4)]
distancia_total = 0

for i in range(len(posiciones)-1):
    x1, y1 = posiciones[i]
    x2, y2 = posiciones[i+1]
    distancia_total += math.sqrt((x2-x1)**2 + (y2-y1)**2)

print("Distancia total recorrida:", distancia_total)
```

## 5) Teniendo como base el siguiente codigo

```python
biblioteca = {
    "libros": [
        {"titulo": "1984", "autor": "George Orwell", "prestado": False},
        {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "prestado": True},
    ],
    "usuarios": ["Ana", "Luis", "María"]
}

# Cantidad de libros prestados y disponibles
prestados = sum(1 for l in biblioteca["libros"] if l["prestado"])
disponibles = len(biblioteca["libros"]) - prestados

print("Prestados:", prestados)
print("Disponibles:", disponibles)

biblioteca["libros"].append({"titulo": "Fahrenheit 451", "autor": "Ray Bradbury", "prestado": False})

for libro in biblioteca["libros"]:
    if libro["titulo"] == "1984":
        libro["prestado"] = True

print(biblioteca)
```
