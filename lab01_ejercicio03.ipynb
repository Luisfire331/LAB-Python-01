## 1) Teniendo en cuenta el siguiente codigo





```python
import requests


url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
usuarios = response.json()

usuarios_biz = []

for u in usuarios:
    if u["email"].endswith(".biz"):
        usuarios_biz.append(u["name"])

print("Usuarios con correo .biz:")
for nombre in usuarios_biz:
    print("-", nombre)
```

## 2) Teniendo en cuenta el siguiente codigo


```python
import requests

nombre_pokemon = input("Ingrese el nombre de un PokÃ©mon: ").lower()

url = f"https://pokeapi.co/api/v2/pokemon/{nombre_pokemon}"
respuesta = requests.get(url)

if respuesta.status_code == 200:
    datos = respuesta.json()

    pokemon = {
        "nombre": datos["name"],
        "altura": datos["height"],
        "peso": datos["weight"],
        "tipos": [t["type"]["name"] for t in datos["types"]],
        "habilidades": [h["ability"]["name"] for h in datos["abilities"]],
    }


    print("\nðŸ“˜ InformaciÃ³n del PokÃ©mon")
    print(f"Nombre: {pokemon['nombre'].capitalize()}")
    print(f"Altura: {pokemon['altura']}")
    print(f"Peso: {pokemon['peso']}")
    print(f"Tipos: {', '.join(pokemon['tipos'])}")


    if len(pokemon["tipos"]) > 1:
        print(" PokÃ©mon de tipo mixto")
    else:
        print(" PokÃ©mon de un solo tipo")

    if pokemon["peso"] > 100:
        print(" PokÃ©mon de gran tamaÃ±o")
    else:
        print(" PokÃ©mon ligero")

    # Listar habilidades una por una
    print("\n Habilidades:")
    for habilidad in pokemon["habilidades"]:
        print("-", habilidad)

    # FUNCIONALIDAD EXTRA:
    # Mostrar estadÃ­sticas base del PokÃ©mon
    print("\n EstadÃ­sticas base:")
    for stat in datos["stats"]:
        print(f"{stat['stat']['name']}: {stat['base_stat']}")

else:
    print("PokÃ©mon no encontrado. Verifique el nombre ingresado.")
```
