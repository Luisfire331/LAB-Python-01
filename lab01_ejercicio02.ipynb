## 2) Teniendo en cuenta el siguiente codigo


```python
n = int(input("Ingrese n: "))

suma = 0
for i in range(1, n + 1):
    suma += i

print("Suma de los primeros n números:", suma)

for i in range(1, n + 1):
    if i % 2 == 0:
        print("Par:", i)

acumulador_pares = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        acumulador_pares += i

print("Suma de pares:", acumulador_pares)


temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]

promedio = sum(temperaturas) / len(temperaturas)

print("Temperaturas mayores al promedio:")
for t in temperaturas:
    if t > promedio:
        print(t)
```

## 3) Teniendo en cuenta el siguiente codigo


```python
contador = 5
while contador > 0:
    print("Contador:", contador)
    contador -= 1

numero = None

while numero != 0:
    numero = int(input("Ingrese un número (0 para salir): "))

print("Programa terminado.")
```

## ¿Qué parte del código representa el estado del ciclo?

El estado del ciclo es la variable que controla su evolución, en este caso:

* contador en el conteo hacia atrás

* numero en el ciclo que termina al ingresar 0

Estas variables definen cómo avanza el ciclo y cuándo finaliza.

## 4) Teniendo en cuenta el siguiente codigo

```python
suma_pares = 0
suma_impares = 0

for i in range(1, 11):
    if i % 2 == 0:
        suma_pares += i
    else:
        suma_impares += i

estado_final = {
    "suma_pares": suma_pares,
    "suma_impares": suma_impares
}

print(estado_final)
```

## ¿Cómo cambia el estado dentro del bucle?

* En cada iteración, las variables suma_pares y suma_impares se actualizan.

* Esto significa que el estado final es el resultado acumulado de todos los cambios ocurridos dentro del ciclo.

## 5) Teniendo como base el siguiente codigo

```python
productos = []

while True:
    nombre = input("Ingrese producto (o 'fin' para salir): ")

    if nombre == "fin":
        break

    precio = float(input("Precio: "))
    cantidad = int(input("Cantidad: "))

    # Saltar productos inválidos
    if precio < 0 or cantidad < 0:
        print("Datos inválidos. Producto ignorado.")
        continue

    productos.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})

total = 0
for p in productos:
    total += p["precio"] * p["cantidad"]

print("Total del inventario:", total)

# Producto más caro
if productos:
    mas_caro = max(productos, key=lambda x: x["precio"])
    print("Producto más caro:", mas_caro)

print("Estado final del inventario:")
print(productos)
```
